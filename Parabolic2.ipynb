```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True #Permite el uso de LaTeX en pyplot

g=9.81
def ymax(dic):
    return (dic['v0']*np.sin(dic['theta0']))**2/(2*g)
```

Aquí se implementa la función ya definida, y arroja el valor máximo de la altura.

```python
In={'v0':float(input(f"Ingrese valor de la velocidad:\t")),
   'theta0':float(input(f"Ingrese el angulo en radianes:\t"))}
ym=ymax(In)
print(f"La altura máxima es de {ym:.4f} metros")
```

```python
ii=0
th=np.linspace(0,np.pi/2, num=20)
v=2.34
n=len(th)
vals={'v0':2.34,
     'theta0':0}
y=np.zeros(n)
for t in th:
    y[ii]=ymax(vals)
    vals['theta0']=t
    ii+=1
```

```python
plt.plot(th,y,label=f"Trayectoria",color="blue")
plt.xlabel(r"$\theta[Rad]$",fontsize=12)
plt.xlim([min(th),max(th)])
plt.ylabel(r"$y_{max}[m]$", fontsize=12)
plt.title(f"Altura máxima",fontsize=15)
plt.grid(True)
plt.savefig(f"Altmax.pdf")
```
