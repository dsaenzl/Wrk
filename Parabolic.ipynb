```python
#Datos iniciales
import numpy as np
import matplotlib.pyplot as plt

In={'v0':0.0,
   'theta0':0.0}
In['v0']=float(input(f"Ingrese la velocidad inicial:\t"))
In['theta0']=float(input(f"Ingrese un numero entre el 0 y 1/2:\t"))*np.pi
#In['v0']=2.34
#In['theta0']=0.25*np.pi
#print(np.pi)
```

```python
g=9.81
vx=In['v0']*np.cos(In['theta0'])
vy=In['v0']*np.sin(In['theta0'])
t=np.arange(0.01,2*vy/g,0.01)
n=len(t)
x=np.zeros(n+1)
y=np.zeros(n+1)

for i in t:
    ii=int(i*100)
    x[ii]=vx*i
    y[ii]=vy*i-(1/2)*g*(i**2)
    if i>0 and y[ii]<=0:
        break
```

```python
print(f"La altura máxima es:\t{((vy**2)/(2*g))}")
plt.plot(x,y,label=f"Trayectoria",color="blue")
plt.xlabel(f"x[m]")
plt.ylabel(f"y[m]")
plt.title(f"Trayectoria")
plt.grid(True)
```
